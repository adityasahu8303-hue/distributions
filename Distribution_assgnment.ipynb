Ques3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and
55 if the data follows a normal distribution?
Ans--Step 1: Identify the range

Mean (μ) = 50

Standard Deviation (σ) = 5

Range given: 45 to 55

Now check how many standard deviations this range covers:
45=50-5=45
55=50+5=55
So the range 45 to 55 is within ±1 standard deviation from the mean.
Step 2: Apply Empirical Rule
According to the 68–95–99.7 rule:
About 68% of the data lies within ±1 standard deviation of the mean.
Ques4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset?
-- Standardization is the process of converting data values into a common scale so they can be compared easily.

The standardized value is called a Z-score.

A Z-score tells us:

How many standard deviations a value is away from the mean.

📐 Formula of Z-Score
𝑧=𝑥−𝜇𝜎z=σx−μ
	​Where:

x = original value

μ (mu) = mean of the dataset

σ (sigma) = standard deviation

z = standardized value (Z-score)

🔄 How Standardization Transforms Data

When we standardize:

The new mean becomes 0

The new standard deviation becomes 1

The shape of the distribution remains the same

Only the scale changes
Why Standardization is Important.....
1. Compare Different Scales

If one exam is out of 100 and another is out of 500, Z-scores help compare performance fairly.

2.  Detect Outliers

Z-score > +3 or < −3 often indicates an outlier.

3. Required for Many Statistical Methods

Used in:

1. Hypothesis testing

2. Normal distribution probability calculations

3. Machine learning algorithms (like logistic regression, SVM, etc.)

4. Makes Data Unit-Free

After standardization, values no longer depend on original units.

Ques 6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die.
-- A uniform distribution is a probability distribution where all outcomes are equally likely.

When we roll a fair six-sided die, the possible outcomes are:1,2,3,4,5,6
For a fair die :...
p(1)=p(2)=p(3)=p(4)=p(5)=p(6)=1/6
A fair die follows a discrete uniform distribution because:

1.  All outcomes are equally likely
2. There is a fixed number of outcomes
3. Long-run frequencies are equal

So, the uniform distribution is the perfect theoretical model for rolling a fair die.
Ques 7. How does log transformation help in stabilizing variance and making data more normally distributed?

-- Log transformation is used in data analysis to stabilize variance and make a dataset more normally distributed, especially when the data is highly right-skewed.
In many real-world datasets, larger values tend to have greater variability than smaller values.
.. a problem known as heteroscedasticity (non-constant variance). When we apply a log transformation, each value is replaced by its logarithm, which compresses large values more than small ones.
.. This reduces the impact of extreme observations and brings the spread of the data to a more consistent level, thereby stabilizing variance.
-- Additionally, log transformation pulls in the long right tail of a positively skewed distribution, making the shape more symmetric and closer to a normal distribution.
-- This is particularly useful because many statistical methods assume normality and constant variance.
-- However, log transformation can only be applied to positive values, since the logarithm of zero or negative numbers is undefined.
